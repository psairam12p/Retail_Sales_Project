```python
import pandas as pd
```

```python
customers=pd.read_csv(r"D:\New folder\jupyter folder\customers.csv")
suppliers=pd.read_csv(r"D:\New folder\jupyter folder\suppliers.csv")
products=pd.read_csv(r"D:\New folder\jupyter folder\products.csv")
stores=pd.read_csv(r"D:\New folder\jupyter folder\stores.csv")
orders=pd.read_csv(r"D:\New folder\jupyter folder\orders.csv")
order_items=pd.read_csv(r"D:\New folder\jupyter folder\order_items.csv")
```

```python
customers.head()
```

```python
# Convert customer_id to int
customers['customer_id'] = customers['customer_id'].astype(int)
customers['registration_date']=pd.to_datetime(customers['registration_date'])
```

```python
customers.drop('resistration_date',axis=1, inplace = True)
```

```python
customers
```

```python
# Fill missing values in 'city' with mode
customers['city'].fillna(customers['city'].mode()[0], inplace=True)
```

```python
# Drop rows where email is missing (critical column)
customers.dropna(subset=['email'], inplace=True)
```

```python
# === Products ===
# Convert price and cost to float
products['price'] = products['price'].astype(float)
products['cost'] = products['cost'].astype(float)
```

```python
# Fill missing 'cost' with median
products['cost'].fillna(products['cost'].median(), inplace=True)

# Fill missing category with mode
products['category'].fillna(products['category'].mode()[0], inplace=True)
```

```python
# === Orders ===
orders['order_id'] = orders['order_id'].astype(int)
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Fill missing 'total_amount' with mean
orders['total_amount'].fillna(orders['total_amount'].mean(), inplace=True)
```

```python
# === Order Items ===
order_items['quantity'] = order_items['quantity'].astype(int)
order_items['product_id'] = order_items['product_id'].astype(int)

# Drop rows with null quantity or product_id (critical for analysis)
order_items.dropna(subset=['quantity', 'product_id'], inplace=True)
```

```python
# === Stores ===
stores['store_id'] = stores['store_id'].astype(int)

# Fill missing store name with "Unknown"
stores['store_name'].fillna("Unknown", inplace=True)
```

```python
# Assuming these DataFrames already exist:
# customers, suppliers, products, stores, orders, order_items

dataframes = {
    'customers': customers,
    'suppliers': suppliers,
    'products': products,
    'stores': stores,
    'orders': orders,
    'order_items': order_items
}

for name, df in dataframes.items():
    print(f"=== {name.upper()} ===")
    print(df.info())
    print("\n")
```
